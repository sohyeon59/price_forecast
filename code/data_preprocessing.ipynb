```python
import pandas as pd
import datetime
import os

dirnm="C:\PSH\캡스톤디자인\data\배추가격"  # 연도별 배추 가격 데이터 포함
dirl=os.listdir(dirnm)
```

```python
dataframes = []

for file in dirl:
    file_path = os.path.join(dirnm, file)  # 전체 파일 경로 생성
    try:
        data = pd.read_csv(file_path, encoding='cp949')  # utf-8로 시도
        dataframes.append(data)
    except UnicodeDecodeError:
        # utf-8로 실패할 경우 euc-kr로 시도
        data = pd.read_csv(file_path, encoding='euc-kr')
        dataframes.append(data)

dataframes=pd.concat(dataframes,axis=0)
dataframes=dataframes.sort_values(by='구분')

# 날짜 열을 datetime 형식으로 변환
dataframes['구분'] = pd.to_datetime(dataframes['구분'], errors='coerce')  # 변환 불가능한 값은 NaT로 처리

# 2020년부터 마지막 연도까지 데이터 필터링
dataframes_2020_onwards = dataframes[dataframes['구분'].dt.year >= 2020]

# dataframes_2020_onwards.to_csv('./data/일간배추가격.csv', index=False)

dataframes_2020_onwards
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_excel("./data/배추공급.xlsx")
data
```

```python
# '총거래물량' 열의 이름을 '거래단위별 총거래물량'으로 변경
data = data.rename(columns={'총거래물량': '거래단위별 총거래물량'})

# 열 제거
data = data.drop(data.columns[2], axis=1)
data = data.drop(data.columns[3], axis=1)
data = data.drop(data.columns[4:6], axis=1)
data = data.drop(data.columns[6], axis=1)
data
```

```python
# 열 이름  변경
data.columns = ['날짜','거래단위','거래단위별 총거래물량','도매지역','품종','산지']

# 도매지역 열에서 앞의 2글자만 추출하여 새로운 열 생성
two = data['도매지역'].str[:2]
# 두 변수 'two'의 내용으로 'data'의 '도매지역' 열을 업데이트
data['도매지역'] = two
data
```

```python
# '산지' 열의 고유한 값 확인 (중복 없이)
print(data['산지'].unique())
```

```python
# '지역명'에서 '제주도'를 '제주'로, '강원도'를 '강원'으로 변경
data['산지'] = data['산지'].replace({'전라남도': '전남', '강원도': '강원', '전라북도': '전북', '경상북도': '경북', '충청남도': '충남',
                                '충청북도': '충북', '서울특별시':'서울','경기도':'경기','경상남도':'경남','세종시':'충남','대구광역시':'대구',
                                '부산광역시':'부산','인천광역시':'인천','대전광역시':'대전','광주광역시':'광주','울산광역시':'울산','제주도':'제주'})
print(data['산지'].unique())
```

```python
# '산지' 열에서 '중국', '-', '제주', '수입산', '기타국', '필리핀' 값을 제거
data = data[~data['산지'].isin(['중국', '-', '수입산', '기타국', '필리핀'])]
print(data['산지'].unique())
```

```python
# 거래단위별 총거래물량 열에서 'kg' 앞의 글자만 남기기
data['거래단위'] = data['거래단위'].astype(str).str.split('kg').str[0]

# '거래단위' 열과 '거래단위별 총거래물량' 열을 숫자형으로 변환
data['거래단위'] = pd.to_numeric(data['거래단위'], errors='coerce')
data['거래단위별 총거래물량'] = pd.to_numeric(data['거래단위별 총거래물량'], errors='coerce')

# 새로운 '총거래물량' 열을 추가하고, '거래단위'와 '거래단위별 총거래물량' 값을 곱하여 채우기
data['총거래물량'] = data['거래단위'] * data['거래단위별 총거래물량']
data
```

```python
# 거래단위, 거래단위별 총거래물량, 열 제거
data = data.drop(data.columns[1:3], axis=1)

# 열 순서 재정렬
data = data[['날짜', '도매지역', '총거래물량', '품종', '산지']]
data
```

```python
# 월 추출
data['월'] = data['날짜'].dt.month

# 조건에 따라 '배추', '쌈배추', '기타배추'만 변경
def change_cabbage_type(row):
    if row['품종'] in ['배추', '쌈배추', '기타배추', '생채용 배추']:  # 해당 품종만 변경
        if row['월'] in [5, 6]:
            return '봄배추'
        elif row['월'] in [7, 8, 9]:
            return '고냉지배추'
        elif row['월'] in [10, 11, 12]:
            return '가을배추'
        elif row['월'] in [1, 2, 3, 4]:
            return '월동배추'
    return row['품종']  # 조건에 맞지 않는 경우 기존 품종 유지

data['품종'] = data.apply(change_cabbage_type, axis=1)

# 월 컬럼 제거
data = data.drop(columns=['월'])

# 5개 지역 남기기
allowed_regions = ['대전', '광주', '서울', '부산', '대구']
data = data[data['도매지역'].isin(allowed_regions)]

# 날짜와 도매지역으로 그룹화하여 총거래물량 합산
data = data.groupby(['날짜', '산지', '품종', '도매지역'], as_index=False).agg({'총거래물량': 'sum'})
```

```python
# 도시별 결측값 선형 보간

import pandas as pd

# 도시 목록
cities = ['서울', '부산', '대구', '대전', '광주']

# 결과를 저장할 빈 데이터프레임
merged_results = pd.DataFrame()

# 반복문으로 각 도시 데이터 처리
for city in cities:
    # 해당 도시 데이터 필터링
    city_data = df[df['도매지역'] == city].copy()
    
    # 날짜를 datetime 형식으로 변환
    city_data['날짜'] = pd.to_datetime(city_data['날짜'])
    
    # 날짜 기준으로 데이터 정렬
    city_data = city_data.sort_values(by='날짜')

    # 날짜 범위 생성 (시작일과 끝일 사이의 모든 날짜)
    full_date_range = pd.date_range(start=city_data['날짜'].min(), end=city_data['날짜'].max(), freq='D')

    # '날짜'를 인덱스로 설정하여 데이터프레임 확장
    city_data.set_index('날짜', inplace=True)

    # 날짜 범위에 맞게 인덱스 재설정 (전체 날짜 범위에 맞게 데이터프레임 확장)
    city_data = city_data.reindex(full_date_range)

    # 결측치를 앞뒤값의 평균으로 보간 (선형 보간법)
    city_data = city_data.interpolate(method='linear')
    
    # '도매지역' 결측치는 앞의 값으로 채우기 (ffill)
    city_data['도매지역'] = city_data['도매지역'].ffill()

    # 인덱스를 '날짜'로 리셋
    city_data.reset_index(inplace=True)
    city_data.rename(columns={'index': '날짜'}, inplace=True)

    # 결과를 하나의 데이터프레임에 병합
    merged_results = pd.concat([merged_results, city_data], axis=0)

merged_results
```

```python
# 서울 이상치 처리

seoul = merged_results[merged_results['도매지역'] == '서울'].copy()
# 상위 1% 임계값 계산
threshold = seoul['총거래물량'].quantile(0.99)

# 이상치에 대한 마스크 생성
outliers_mask = seoul['총거래물량'] > threshold

# 이상치를 NaN으로 변환
seoul.loc[outliers_mask, '총거래물량'] = None

# 선형 보간법 적용
seoul['총거래물량'] = seoul['총거래물량'].interpolate(method='linear')
seoul
```

```python
# 광주 이상치 처리

gwangju = merged_results[merged_results['도매지역'] == '광주'].copy()
# 상위 1% 임계값 계산
threshold = gwangju['총거래물량'].quantile(0.99)

# 이상치에 대한 마스크 생성
outliers_mask = gwangju['총거래물량'] > threshold

# 이상치를 NaN으로 변환
gwangju.loc[outliers_mask, '총거래물량'] = None

# 선형 보간법 적용
gwangju['총거래물량'] = gwangju['총거래물량'].interpolate(method='linear')
gwangju
```

```python
filtered_results = merged_results[~merged_results['도매지역'].isin(['서울', '광주'])]  # 서울, 광주 제외
updated_results = pd.concat([filtered_results, seoul, gwangju], ignore_index=True)  # 이상치 처리한 서울, 광주

updated_results = updated_results.sort_values(by=['도매지역', '날짜']).reset_index(drop=True)  # 도매지역, 날짜 순으로 정렬
```

```python
import pandas as pd

data = pd.read_csv('./data/일간배추가격.csv')

# 지역 데이터를 "지역"과 "가격" 열로 변환
df = data.melt(
    id_vars=['구분', '평균', '등락율'],  # 고정할 열
    value_vars=['서울', '부산', '대구', '광주', '대전'],  # 녹일 열
    var_name='지역',  # 새로운 열 이름 (기존 열 이름)
    value_name='가격'  # 새로운 값 열 이름
)

df['가격'] = df['가격'].replace('-', '0')  # '-'을 '0'으로 변경
df['가격'] = df['가격'].str.replace(',', '').astype(int)
df.drop(['평균', '등락율'], axis=1, inplace=True)
df.columns = ['날짜', '지역', '가격']

가격 = df
가격
```
