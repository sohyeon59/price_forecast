# 가격예측 지역별

```python
import pandas as pd

df = pd.read_csv('price_model최최종.csv')
df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Conv1D, MaxPooling1D, SimpleRNN
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

regions = df['지역'].unique()
results = {}

# 지역별 데이터 처리
for region in regions:
    print(f"Processing for region: {region}")
    data = df[df['지역'] == region]
    data = data[['배추가격', '예측공급량', "물가지수", "배추 수입량", "무가격"]]

    # Feature & Target 분리
    target = data['배추가격']
    features = data.drop(columns=['배추가격'])

    # 데이터 전처리 (스케일링)
    scaler = MinMaxScaler()
    scaled_features = scaler.fit_transform(features)
    scaled_target = scaler.fit_transform(target.values.reshape(-1, 1))

    # 시계열 데이터 생성
    def create_sequences(data, target, seq_len):
        X, y = [], []
        for i in range(seq_len, len(data)):
            X.append(data[i-seq_len:i])
            y.append(target[i])
        return np.array(X), np.array(y)

    seq_len = 30
    X, y = create_sequences(scaled_features, scaled_target, seq_len)

    # Train-Test Split
    train_size = int(len(y) * 0.7)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    # ----- LSTM 모델 -----
    lstm_model = Sequential([
        LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
        LSTM(50, return_sequences=False),
        Dense(1)
    ])
    lstm_model.compile(optimizer='adam', loss='mse')
    lstm_model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0)
    lstm_pred = lstm_model.predict(X_test)

    # ----- CNN-LSTM 모델 -----
    cnn_lstm_model = Sequential([
        Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])),
        MaxPooling1D(pool_size=2),
        LSTM(50, return_sequences=False),
        Dense(1)
    ])
    cnn_lstm_model.compile(optimizer='adam', loss='mse')
    cnn_lstm_model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0)
    cnn_lstm_pred = cnn_lstm_model.predict(X_test)

    # ----- RNN 모델 -----
    rnn_model = Sequential([
        SimpleRNN(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
        SimpleRNN(50, return_sequences=False),
        Dense(1)
    ])
    rnn_model.compile(optimizer='adam', loss='mse')
    rnn_model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0)
    rnn_pred = rnn_model.predict(X_test)

    # ----- CNN-RNN 모델 -----
    cnn_rnn_model = Sequential([
        Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])),
        MaxPooling1D(pool_size=2),
        SimpleRNN(50, return_sequences=False),
        Dense(1)
    ])
    cnn_rnn_model.compile(optimizer='adam', loss='mse')
    cnn_rnn_model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0)
    cnn_rnn_pred = cnn_rnn_model.predict(X_test)

    # ----- 평가 -----
    def evaluate_model(y_test, predictions):
        mae = mean_absolute_error(scaler.inverse_transform(y_test), scaler.inverse_transform(predictions))
        mse = mean_squared_error(scaler.inverse_transform(y_test), scaler.inverse_transform(predictions))
        rmse = np.sqrt(mse)
        r2 = r2_score(scaler.inverse_transform(y_test), scaler.inverse_transform(predictions))
        return mae, mse, rmse, r2

    lstm_metrics = evaluate_model(y_test, lstm_pred)
    cnn_lstm_metrics = evaluate_model(y_test, cnn_lstm_pred)
    rnn_metrics = evaluate_model(y_test, rnn_pred)
    cnn_rnn_metrics = evaluate_model(y_test, cnn_rnn_pred)

    # 결과 저장
    results[region] = {
        'LSTM': {'MAE': lstm_metrics[0], 'MSE': lstm_metrics[1], 'RMSE': lstm_metrics[2], 'R^2': lstm_metrics[3]},
        'CNN-LSTM': {'MAE': cnn_lstm_metrics[0], 'MSE': cnn_lstm_metrics[1], 'RMSE': cnn_lstm_metrics[2], 'R^2': cnn_lstm_metrics[3]},
        'RNN': {'MAE': rnn_metrics[0], 'MSE': rnn_metrics[1], 'RMSE': rnn_metrics[2], 'R^2': rnn_metrics[3]},
        'CNN-RNN': {'MAE': cnn_rnn_metrics[0], 'MSE': cnn_rnn_metrics[1], 'RMSE': cnn_rnn_metrics[2], 'R^2': cnn_rnn_metrics[3]},
    }

    # ----- 시각화 -----
    plt.figure(figsize=(12, 6))
    plt.plot(scaler.inverse_transform(y_test), label="Actual", color="blue")
    plt.plot(scaler.inverse_transform(lstm_pred), label="LSTM Prediction", color="red")
    plt.plot(scaler.inverse_transform(cnn_lstm_pred), label="CNN-LSTM Prediction", color="green")
    plt.plot(scaler.inverse_transform(rnn_pred), label="RNN Prediction", color="orange")
    plt.plot(scaler.inverse_transform(cnn_rnn_pred), label="CNN-RNN Prediction", color="purple")
    plt.title(f"Actual vs Predicted Prices for {region}")
    plt.legend()
    plt.show()
```

```python
import pandas as pd

# 지역별 성능 지표 결과를 담을 리스트 생성
performance_data = []

# 지역별 성능 지표를 리스트에 추가
for region, metrics in results.items():
    row = {'지역': region}
    for model, metrics_values in metrics.items():
        row[f'{model}_MAE'] = metrics_values['MAE']
        row[f'{model}_MSE'] = metrics_values['MSE']
        row[f'{model}_RMSE'] = metrics_values['RMSE']
        row[f'{model}_R^2'] = metrics_values['R^2']
    performance_data.append(row)

# 성능 지표 DataFrame 생성
performance_df = pd.DataFrame(performance_data)

# 행렬을 변환하여 모델과 성능 지표가 행, 지역명이 열로 변환된 DataFrame 생성
performance_df_transposed = performance_df.set_index('지역').transpose()

# R^2에 해당하는 행만 빨간색으로 스타일링하는 함수 정의
def highlight_r2(row):
    return ['color: red' if 'R^2' in row.name else '' for _ in row]

# Styler를 사용하여 R^2 행에 스타일 적용 및 소수점 자리수 지정
styled_df = (
    performance_df_transposed.style
    .apply(highlight_r2, axis=1)
    .format(precision=3)  # 소수점 자리수 설정
)

styled_df
```

## 가격예측 글로벌

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Conv1D, MaxPooling1D, SimpleRNN
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import platform

df = pd.read_csv('price_model최최종.csv')

# Feature & Target 분리
target = df['배추가격']

# 날짜 처리: 문자열 데이터를 숫자형으로 변환하거나 제거
if '날짜' in df.columns:
    df['날짜'] = pd.to_datetime(df['날짜'])
    df['날짜_순서'] = (df['날짜'] - df['날짜'].min()).dt.days  # 날짜를 숫자로 변환
    features = df.drop(columns=['배추가격', '날짜'])  # '날짜' 대신 '날짜_순서'를 사용
else:
    features = df.drop(columns=['배추가격'])

# 원-핫 인코딩
features = pd.get_dummies(features, columns=['지역'], drop_first=True)

# 데이터 전처리 (스케일링)
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(features)
scaled_target = scaler.fit_transform(target.values.reshape(-1, 1))

# 시계열 데이터 생성
def create_sequences(data, target, seq_len):
    X, y = [], []
    for i in range(seq_len, len(data)):
        X.append(data[i-seq_len:i])
        y.append(target[i])
    return np.array(X), np.array(y)

seq_len = 30
X, y = create_sequences(scaled_features, scaled_target, seq_len)

# Train-Test Split
train_size = int(len(y) * 0.7)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# ----- 글로벌 LSTM 모델 -----
lstm_model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    LSTM(50, return_sequences=False),
    Dense(1)
])
lstm_model.compile(optimizer='adam', loss='mse')
lstm_model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0)
lstm_pred = lstm_model.predict(X_test)

# ----- 글로벌 CNN-LSTM 모델 -----
cnn_lstm_model = Sequential([
    Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])),
    MaxPooling1D(pool_size=2),
    LSTM(50, return_sequences=False),
    Dense(1)
])
cnn_lstm_model.compile(optimizer='adam', loss='mse')
cnn_lstm_model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0)
cnn_lstm_pred = cnn_lstm_model.predict(X_test)

# ----- 글로벌 RNN 모델 -----
rnn_model = Sequential([
    SimpleRNN(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    SimpleRNN(50, return_sequences=False),
    Dense(1)
])
rnn_model.compile(optimizer='adam', loss='mse')
rnn_model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0)
rnn_pred = rnn_model.predict(X_test)

# ----- 글로벌 CNN-RNN 모델 -----
cnn_rnn_model = Sequential([
    Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])),
    MaxPooling1D(pool_size=2),
    SimpleRNN(50, return_sequences=False),
    Dense(1)
])
cnn_rnn_model.compile(optimizer='adam', loss='mse')
cnn_rnn_model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0)
cnn_rnn_pred = cnn_rnn_model.predict(X_test)

# ----- 평가 -----
def evaluate_model(y_test, predictions):
    mae = mean_absolute_error(scaler.inverse_transform(y_test), scaler.inverse_transform(predictions))
    mse = mean_squared_error(scaler.inverse_transform(y_test), scaler.inverse_transform(predictions))
    rmse = np.sqrt(mse)
    r2 = r2_score(scaler.inverse_transform(y_test), scaler.inverse_transform(predictions))
    return mae, mse, rmse, r2

lstm_metrics = evaluate_model(y_test, lstm_pred)
cnn_lstm_metrics = evaluate_model(y_test, cnn_lstm_pred)
rnn_metrics = evaluate_model(y_test, rnn_pred)
cnn_rnn_metrics = evaluate_model(y_test, cnn_rnn_pred)

# ----- 성능평가표 생성 -----
results = {
    "LSTM": {"MAE": lstm_metrics[0], "MSE": lstm_metrics[1], "RMSE": lstm_metrics[2], "R^2": lstm_metrics[3]},
    "CNN-LSTM": {"MAE": cnn_lstm_metrics[0], "MSE": cnn_lstm_metrics[1], "RMSE": cnn_lstm_metrics[2], "R^2": cnn_lstm_metrics[3]},
    "RNN": {"MAE": rnn_metrics[0], "MSE": rnn_metrics[1], "RMSE": rnn_metrics[2], "R^2": rnn_metrics[3]},
    "CNN-RNN": {"MAE": cnn_rnn_metrics[0], "MSE": cnn_rnn_metrics[1], "RMSE": cnn_rnn_metrics[2], "R^2": cnn_rnn_metrics[3]}
}

# Create a DataFrame with models as rows and metrics as columns
performance_data = []
for model, metrics_values in results.items():
    row = [metrics_values["MAE"], metrics_values["MSE"], metrics_values["RMSE"], metrics_values["R^2"]]
    performance_data.append(row)

performance_df = pd.DataFrame(performance_data, columns=["MAE", "MSE", "RMSE", "R^2"], index=results.keys())
performance_df
print(performance_df)


# ----- 시각화 -----
plt.figure(figsize=(12, 6))
plt.plot(scaler.inverse_transform(y_test), label="Actual", color="blue")
plt.plot(scaler.inverse_transform(lstm_pred), label="LSTM Prediction", color="red")
plt.plot(scaler.inverse_transform(cnn_lstm_pred), label="CNN-LSTM Prediction", color="green")
plt.plot(scaler.inverse_transform(rnn_pred), label="RNN Prediction", color="orange")
plt.plot(scaler.inverse_transform(cnn_rnn_pred), label="CNN-RNN Prediction", color="purple")
plt.title("Actual vs Predicted Cabbage Prices (Global Model)")
plt.xlabel("Time (Days)")  # X축: 시간 (일 단위)
plt.ylabel("Cabbage Prices (₩)")  # Y축: 배추 가격 (₩)
plt.legend()
plt.grid(True)
plt.show()

# ----- 결과 출력 -----
print("LSTM Metrics:", lstm_metrics)
print("CNN-LSTM Metrics:", cnn_lstm_metrics)
print("RNN Metrics:", rnn_metrics)
print("CNN-RNN Metrics:", cnn_rnn_metrics)
# ----- 시각화 -----
plt.figure(figsize=(12, 6))
plt.plot(scaler.inverse_transform(y_test), label="Actual", color="blue")
plt.plot(scaler.inverse_transform(lstm_pred), label="LSTM Prediction", color="red")
plt.plot(scaler.inverse_transform(cnn_lstm_pred), label="CNN-LSTM Prediction", color="green")
plt.plot(scaler.inverse_transform(rnn_pred), label="RNN Prediction", color="orange")
plt.title("Actual vs Predicted Cabbage Prices (Global Model)")
plt.legend()
plt.show()
```
